```python
test_set=np.array([[7.2,3.6,5.1,2.5]])
test_set
```

```python
test_set.shape
```

### Distance measurement of test data from each given data point.

```python
d1=X-test_set
d1
```

```python
d2=(d1**2).sum(axis=1)
```

```python
dist=np.sqrt(d2)
dist=dist[:,np.newaxis]
dist
```

### Stacking distance of each data points(given) from test data(unseen) with its corresponding label. 

```python
pred_M=np.hstack((Y,dist))
```

```python
pred_M
```

```python
Result=pd.DataFrame(pred_M)
Result
```

### Sorting by distance column ( here-column 1) so that nearest points from the test data come first

```python
final=Result.sort_values(by=[1])
final
```

Note:K=number of neighbour to be selected is taken as square root of total number of sample in data set. Here sqare root of 150 sample~13.Finally categories assignment or classification of unseend data is done based on majority of vote among those 13 neaighboour's class/type.
